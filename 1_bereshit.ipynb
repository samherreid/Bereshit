${^ד}$ויקרא אלהים  לאור **יום${^ב}$** ולחשך קרא **לילה${^ב}$** ויהי־ערב ויהי־בקר **יום אחד**   

```python
import numpy as np
import inspect
import re
import concurrent.futures

a = [(4, 5), (7, 3), (6, 6)]
```

```python
def day1(a):
    return [x * y for x, y in a]

def day2(a):
    return [2 * (x + y) for x, y in a]

def main():
    result = []

    for func in [day1, day2]:
        source = inspect.getsource(func)
        match = re.search(r'return\s*\[\s*(.+?)\s*for', source)
        equation = match.group(1)
        
        computed_results = func(a)
        
        items = [func.__name__, equation] + computed_results
        if computed_results:
            items.append('Good')
        else:
            items.append('Not good')
        
        result.extend(items)
        
    print("Result:", result)

if __name__ == "__main__":
    main()
```

```python
def day1(a):
    eq = 'x * y'
    results = [int(round(np.imag((x + 1j * y)**2) / 2)) for x, y in a]
    return eq, results

def day2(a):
    eq = '2 * (x + y)'
    results = [int(round(2 * np.real(((x + 1j * y) + (y + 1j * x)) / (1 + 1j)))) for x, y in a]
    return eq, results

def main():
    result = []

    for func in [day1, day2]:
        eq, results = func(a)

        items = [func.__name__, eq] + results
        if results:
            items.append('Good')
        else:
            items.append('Not good')
        result.extend(items)
        
    print("Result:", result)

if __name__ == "__main__":
    main()
```

```python
def day1(a):
    eq = 'x * y'
    dummy = 0
    for i in range(10**7):
        dummy += i % 10
    results = [x * y for x, y in a]
    return eq, results

def day2(a):
    eq = '2 * (x + y)'
    results = [2 * (x + y) for x, y in a]
    return eq, results

def main():
    result = []
    funcs = [day1, day2]
    
    with concurrent.futures.ThreadPoolExecutor() as executor:
        future_to_func = {executor.submit(func, a): func for func in funcs}
        for future in concurrent.futures.as_completed(future_to_func):
            func = future_to_func[future]
            eq, results = future.result()
            items = [func.__name__, eq] + results
            if results:
                items.append('Good')
            else:
                items.append('Not good')
            result.extend(items)
    
    print("Result:", result)

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm

x = np.linspace(-np.pi, np.pi, 220)
y = np.linspace(-np.pi, np.pi, 220)
X, Y = np.meshgrid(x, y)

Z1 = np.sin(X) * np.sin(Y)
Z2 = X * (X**2 + Y**2)

def plot_surface(X, Y, Z, elev=28, azim=-55):
   fig = plt.figure(figsize=(8, 6))
   ax = fig.add_subplot(111, projection='3d')

   norm = TwoSlopeNorm(vmin=Z.min(), vcenter=0.0, vmax=Z.max())
   surf = ax.plot_surface(
       X, Y, Z,
       cmap='RdBu_r',
       norm=norm,
       linewidth=0,
       antialiased=True
   )

   ax.view_init(elev=elev, azim=azim)
   plt.tight_layout()
   plt.show()

plot_surface(X, Y, Z1)
plot_surface(X, Y, Z2)
```
